```python
import numpy as np
import matplotlib.pyplot as plt
def gaussian(x, mu, sig):
    return np.exp(-(np.power((x - mu),2)) / (2 * np.power(sig,2)))
def perturbate():
    val = np.random.normal(1, 0.1)
    #val = 2 - val if val > 1 else val
    return val
```

```python
x_values = np.array([0.2, 0.4, 0.7]).reshape(-1,1)
u = np.arange(0,1,0.001)
```

```python
#alpha_k = fmax(1 - x[k]/u[k],0) * u[k]
#a = np.arange(0,1,0.01)
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'cm'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
matplotlib.rcParams['font.size'] = 35
improvement_vect = np.array([])
fig, ax = plt.subplots(1,2, dpi=200, figsize=(19,9))
for x in x_values:
    a = np.maximum((u-x),0)
    #improvement_vect = np.append(improvement_vect, gaussian(a, (1-x)/2, (1-x)/5))
    ax[0].plot(u, gaussian(a, (1-x)/2, (1-x)/8), label=f'$x_i(k)$: %.2f'%x, linewidth=4)
    ax[1].plot(a, gaussian(a, (1-x)/2, (1-x)/8), label=f'$x_i(k)$: %.2f'%x, linewidth=4)
ax[0].set_ylabel('$\delta_i(x,u)$')
ax[0].set_xlabel('$u_i$')
ax[0].legend(loc='best', fontsize=25)
ax[0].margins(x=0)
ax[0].grid(True)
ax[1].set_ylabel('$\delta_i(x,u)$')
ax[1].set_xlabel(r'$\alpha_i=max(u_i-x_i,0)$')
ax[1].legend(loc='best', fontsize=25)
ax[1].margins(x=0)
ax[1].grid(True)
#ax[0].set_ylim([0.04, 1.1])
#ax[1].set_ylim([0.04, 1.1])
ax[0].set_title('(a)', pad=10)
ax[1].set_title('(b)', pad=10)
fig.tight_layout()
```
